```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text=str(text).lower()
  text=re.sub(r'[^a-z\s]', '', text)#removing puncuation
  words=text.split()

  words=[word for word in words if word not in stop_words]
  words=[lemmatizer.lemmatize(word) for word in words]

  return ' '.join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df=df.reset_index(drop=True)
```

```python
df['title'] = df['title'].str.lower()
df.head()
```

```python
indices=pd.Series(df.index,index=df['title'])
indices
```

```python
df = df.reset_index(drop=True)
```

```python
#vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
#using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

```

```python
def recommend(title,n=10):
  if title not in indices:
    return ["Movie not found"]

  idx=indices[title]
  if isinstance(idx, (list, tuple)) or hasattr(idx, "__len__"):
        idx = idx[0]
  sim_score=cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_index=sim_score.argsort()[::-1][1:n+1]
  similar_movies=df.iloc[similar_index]
  return df['title'].iloc[similar_index]
```

```python
recommend('jumanji')
```

```python
import pickle
pickle.dump(tfidf_matrix,open('tfidf_matrix.pkl','wb'))
pickle.dump(df,open('df.pkl','wb'))
pickle.dump(indices,open('indices.pkl','wb'))
pickle.dump(tfidf,open('tfidf.pkl','wb'))
pickle.dump(recommend,open('recommend.pkl','wb'))
```
